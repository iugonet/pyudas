A collection of examples for the `iugonet` functions on top of pySPEDAS, grouped by function. The package is imported as `iugonet`.

# Tools — statistical analysis package

Ports of the UDAS `statistical_package` (`iugonet/tools`): spectral analysis, correlation, trend and
difference tests, interpolation and change-point detection. Routines whose name starts with `u` take
tplot variable names; the others take plain arrays, so they can be combined with anything in NumPy.

The examples below all run on the same real event — the **Halloween storm of 29 October 2003**
(Dst minimum −383 nT) — using 1-minute geomagnetic field data from WDC Kyoto for
**KAK** (Kakioka) and **MMB** (Memambetsu), about 900 km apart.

### Data used in this section

```python
iugonet.gmag_wdc(trange=['2003-10-29', '2003-10-30'], site='kak mmb', resolution='min')
iugonet.gmag_wdc(trange=['2003-10-29', '2003-10-30'], site='dst')

DAY = time_double('2003-10-29')

def day_component(varname, label='H [nT]'):
    """One day of a WDC station, picked by component label.

    WDC minute files are monthly, so a one-day trange still returns the whole month.
    The columns are ordered D, H, Z, F -- pick by label rather than by column index.
    """
    d = get_data(varname)
    names = get_data(varname, metadata=True)['plot_options']['yaxis_opt']['legend_names']
    t = np.asarray(d.times)
    i = np.searchsorted(t, DAY)
    return t[i:i + 1440], np.asarray(d.y)[i:i + 1440, names.index(label)]

for site in ('kak', 'mmb'):
    t, h = day_component('wdc_mag_%s_1min' % site)
    store_data('%s_H' % site, data={'x': t, 'y': h})
    options('%s_H' % site, 'ytitle', site.upper() + ' H')
    options('%s_H' % site, 'ysubtitle', '[nT]')

# Dst comes from a monthly file too, so pin the plot to the one day we care about.
timespan('2003-10-29', 1, keyword='days')
tplot(['kak_H', 'mmb_H', 'wdc_mag_dst'])
```

## `iugonet.ustrans_pwrspc`

### ustrans_pwrspc_kak

```python
# S-transform (Stockwell) dynamic power spectrum. `removeedge` fits and removes a quadratic
# first -- without it the 30000 nT DC level dominates and nothing else is visible.
# minfreq/maxfreq are frequency *indices*: with 1440 samples at 60 s, index k is a period of
# 1440/k minutes, so 12..360 keeps periods of 4..120 min.
iugonet.ustrans_pwrspc('kak_H', 1, samplingrate=60.0, abs=True, removeedge=True,
                       minfreq=12, maxfreq=360)
options('kak_H_stpwrspc', 'ylog', 1)
options('kak_H_stpwrspc', 'ytitle', 'S-transform')
options('kak_H_stpwrspc', 'ysubtitle', 'Period [s]')
options('kak_H_stpwrspc', 'ztitle', '[nT]')
tplot(['kak_H', 'kak_H_stpwrspc'])
```

## `iugonet.uspec_coh`

### uspec_coh_kak_mmb

```python
# Coherence and phase difference between two stations. `width` smooths the periodogram;
# without it the coherence is identically 1 at every frequency.
res = iugonet.uspec_coh('kak_H', 'mmb_H', deltat=60.0, width=10)

f = np.asarray(res['freq'])
cxy = np.asarray(res['result']['cxy'])
ok = f > 0                                   # f[0] is DC, so 1/f is inf there

plt.figure(figsize=(9, 3.5))
plt.semilogx(1.0 / f[ok] / 60.0, cxy[ok], lw=0.8)
plt.axhline(res['gama'], color='r', ls='--',
            label='confidence level = %.3f' % res['gama'])
plt.xlabel('Period [min]'); plt.ylabel('Coherence')
plt.title('KAK vs MMB, 2003-10-29 (max coherence = %.4f)' % res['max_cxy'])
plt.legend(); plt.grid(alpha=0.3); plt.show()
```

## `iugonet.coherence_analysis`

### coherence_analysis_kak_mmb

```python
# Same quantity as uspec_coh, but with a different confidence-interval formula
# (g2 depends on the series length) and the smoothing window fixed to 'hanning'.
res = iugonet.coherence_analysis('kak_H', 'mmb_H', deltat=60.0, width=10)
print('max coherence =', res['max_cxy'])
print('main period   =', np.asarray(res['main_period']).ravel()[:3], '[s]')
```

## `iugonet.ucross_cor`

### ucross_cor_kak_mmb

```python
# Cross-correlation over a lag range, with a t-test on the peak.
cc = iugonet.ucross_cor('kak_H', 'mmb_H', low_lag=-60, high_lag=60)

lags = np.arange(-60, 61)
plt.figure(figsize=(9, 3.5))
plt.plot(lags, np.asarray(cc['cross_cor']), lw=1)
plt.xlabel('Lag [min]'); plt.ylabel('Cross-correlation')
plt.title('KAK vs MMB (max %.3f at lag %d min)'
          % (np.asarray(cc['max_corr']).ravel()[0], np.asarray(cc['max_zure']).ravel()[0]))
plt.grid(alpha=0.3); plt.show()
```

## `iugonet.uchange_point_checker`

### uchange_point_checker_kak

```python
# Two-segment (broken-stick) regression: the F statistic peaks where the series changes slope.
cp = iugonet.uchange_point_checker('kak_H')
fc1 = np.asarray(cp['Fc1'])

plt.figure(figsize=(9, 3.5))
plt.plot(np.arange(2, 2 + fc1.size) / 60.0, fc1, lw=0.8)
plt.axvline(cp['change_point'] / 60.0, color='r', ls='--',
            label='change point = %.2f h UT' % (cp['change_point'] / 60.0))
plt.xlabel('Hour UT'); plt.ylabel('F statistic')
plt.title('uchange_point_checker on KAK H, 2003-10-29')
plt.legend(); plt.grid(alpha=0.3); plt.show()
```

## `iugonet.utrend_test`

### utrend_test_kak

```python
# Rank-based trend test. |Z| above the threshold means a significant trend; the sign of Z
# gives its direction. Here it tracks the storm phase: rising before onset, falling in the
# main phase.
t, h = day_component('wdc_mag_kak_1min')
for label, a, b in [('00-04 UT  (before onset)', 0, 240),
                    ('06-10 UT  (main phase onset)', 360, 600),
                    ('20-24 UT  (main phase)', 1200, 1440)]:
    store_data('win', data={'x': t[a:b], 'y': h[a:b]})
    print('--- %s   dH = %+.0f nT' % (label, h[b - 1] - h[a]))
    iugonet.utrend_test('win', sl=0.05)
```

## `iugonet.udifference_test`

### udifference_test_quiet_vs_storm

```python
# Are two samples drawn from the same distribution? test_sel=0 runs a normality test first
# and picks Welch (normal) or Mann-Whitney (otherwise); 1 forces Welch, 2 forces Mann-Whitney.
# Compare a quiet day with the storm day -- both come from the same monthly WDC file.
d = get_data('wdc_mag_kak_1min')
t = np.asarray(d.times)
names = get_data('wdc_mag_kak_1min', metadata=True)['plot_options']['yaxis_opt']['legend_names']
q = np.searchsorted(t, time_double('2003-10-25'))
store_data('kak_quiet', data={'x': t[q:q + 1440], 'y': np.asarray(d.y)[q:q + 1440, names.index('H [nT]')]})

iugonet.udifference_test('kak_quiet', 'kak_H', test_sel=0)
```

## `iugonet.gmag_wdc_xyz`

### gmag_wdc_xyz_kak

```python
# H/D/Z -> X/Y/Z for WDC stations already loaded by gmag_wdc. The components are found by
# their labels, not by column order (a WDC row is D, H, Z, F).
iugonet.gmag_wdc_xyz(site='kak', resolution='min')

# Check the conversion: X = H cos D, Y = H sin D (D is in degrees in the WDC file).
t, h = day_component('wdc_mag_kak_1min', 'H [nT]')
_, dec = day_component('wdc_mag_kak_1min', 'D [deg]')
d = get_data('wdc_mag_kak_1min_xyz')
names = get_data('wdc_mag_kak_1min_xyz', metadata=True)['plot_options']['yaxis_opt']['legend_names']
i = np.searchsorted(np.asarray(d.times), DAY)
xyz = np.asarray(d.y)[i:i + 1440]
print('max|X - H cos D| =', np.abs(xyz[:, 0] - h * np.cos(np.radians(dec))).max(), 'nT')
print('max|Y - H sin D| =', np.abs(xyz[:, 1] - h * np.sin(np.radians(dec))).max(), 'nT')

# The absolute field is ~30000 nT while the storm is a few hundred nT, so plot the
# deviation from each component's daily mean to make the storm visible.
plt.figure(figsize=(9, 3.5))
for k, nm in enumerate(names):
    plt.plot((t - DAY) / 3600.0, xyz[:, k] - xyz[:, k].mean(), lw=0.8, label=nm)
plt.xlabel('Hour UT'); plt.ylabel('Deviation from daily mean [nT]')
plt.title('gmag_wdc_xyz — KAK, 2003-10-29'); plt.legend(); plt.grid(alpha=0.3); plt.show()
```

## Numeric cores

The routines above are tplot wrappers around these. They take plain arrays, so they work on
anything -- not only tplot variables.

### s_trans / hilbert_trans / gaussian_window

```python
t, h = day_component('wdc_mag_kak_1min')

r = iugonet.s_trans(h, 1, samplingrate=60.0, abs=True, removeedge=True, minfreq=12, maxfreq=360)
st, freq = np.asarray(r['st']), np.asarray(r['freq'])
print('s_trans        ->', st.shape, '(time, freq)')

print('hilbert_trans  ->', np.asarray(iugonet.hilbert_trans(h)).shape, '(analytic signal)')
print('gaussian_window->', np.asarray(iugonet.gaussian_window(64, 4.0)).shape)

plt.figure(figsize=(9, 3.5))
plt.pcolormesh((t - DAY) / 3600.0, 1.0 / freq / 60.0, st.T, shading='auto', cmap='jet')
plt.yscale('log'); plt.xlabel('Hour UT'); plt.ylabel('Period [min]')
plt.title('s_trans amplitude, KAK H'); plt.colorbar(label='[nT]'); plt.show()
```

### cross_spec / filter_window / idl_filter

```python
t, hk = day_component('wdc_mag_kak_1min')
_, hm = day_component('wdc_mag_mmb_1min')

r = iugonet.cross_spec(hk, hm, deltat=60.0, width=10, window='hanning')
print('cross_spec keys:', list(r))
print('  all tags have the same length:', {k: len(v) for k, v in r.items()})

print('filter_window(11, "hanning") =', np.round(iugonet.filter_window(11, 'hanning'), 4))
print('idl_filter(...) ->', np.asarray(iugonet.idl_filter(hk, 11, 'boxcar')).shape)
```

### c_cor

```python
# Cross-correlation of two series plus each one's autocorrelation, over a lag range.
# (ucross_cor is the tplot wrapper around this, and adds the significance test.)
t, hk = day_component('wdc_mag_kak_1min')
_, hm = day_component('wdc_mag_mmb_1min')
r = iugonet.c_cor(hk, hm, low_lag=-60, high_lag=60)
print('c_cor keys:', list(r))

lags = np.arange(r['low_lag'], r['high_lag'] + 1)
plt.figure(figsize=(9, 3.5))
plt.plot(lags, r['cross_cor'], lw=1, label='cross-correlation KAK x MMB')
plt.plot(lags, r['x_cor'], lw=0.8, ls='--', label='autocorrelation KAK')
plt.plot(lags, r['y_cor'], lw=0.8, ls='--', label='autocorrelation MMB')
plt.xlabel('Lag [min]'); plt.ylabel('Correlation')
plt.title('c_cor, 2003-10-29 (peak %.3f at lag %d min)'
          % (r['cross_cor'].max(), lags[np.argmax(r['cross_cor'])]))
plt.legend(fontsize=8); plt.grid(alpha=0.3); plt.show()
```

### trend_test / welch_test / mann_whitney_test / normality_test / difference_test

```python
t, h = day_component('wdc_mag_kak_1min')
first, second = h[:720], h[720:]          # first vs second half of the storm day

print('trend_test        Z =', iugonet.trend_test(h, quiet=True))
print('normality_test      =', iugonet.normality_test(h, quiet=True), ' (0 = normal, 1 = not normal)')
print('welch_test          =', iugonet.welch_test(first, second, quiet=True), ' (1 = same mean, 0 = differ)')
print('mann_whitney_test   =', iugonet.mann_whitney_test(first, second, quiet=True), ' (3 = same, 2 = differ)')
print('difference_test     =', iugonet.difference_test(first, second, quiet=True))
```
