## 2. Load data and plot

Load NIPR fluxgate magnetometer data (Syowa Station) and plot it with `pyspedas.tplot`.

```python
import iugonet
from pyspedas import tplot

iugonet.gmag_nipr(trange=['2003-10-29', '2003-10-30'], site='syo')
tplot('nipr_mag_syo_1sec')
```

See the README for the full list of load functions. Check a function's arguments with `help(iugonet.gmag_nipr)`.
