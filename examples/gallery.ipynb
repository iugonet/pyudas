A collection of load examples for the `iugonet` functions on top of pySPEDAS, grouped by function. Each cell loads data with example parameters and shows the resulting tplot variables. The package is imported as `iugonet`.

> Some high-resolution datasets use short time ranges to stay light; multi-dimensional model variables (GAIA / Kyushu-GCM 4-D, GPS-TEC 3-D) are listed rather than plotted.

```python
import numpy as np
import iugonet
from pyspedas import tplot, get_data, store_data, options
print('iugonet', iugonet.__version__)
```

## `iugonet.gmag_nipr`

### gmag_nipr_syo

```python
iugonet.gmag_nipr(trange=['2003-10-29', '2003-10-30'], site='syo')
tplot('nipr_mag_*')
```

## `iugonet.kyushugcm`

### kyushugcm_T

```python
iugonet.kyushugcm(trange=['2010-01-01', '2010-01-02'], datatype='T')
d = get_data('kyushugcm_T')
y = np.asarray(d.y)[:, 32, 64, :]
store_data('kyushugcm_T_prof', data={'x': d.times, 'y': y, 'v': np.arange(y.shape[1])})
options('kyushugcm_T_prof', 'spec', 1)
tplot('kyushugcm_T_prof')
```
