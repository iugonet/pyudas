Runnable `iugonet.<function>(...)` examples for **every case** validated bit-identical against IDL UDAS (see the comparison report). Grouped by function; each cell loads the data with the validated parameters and shows the resulting tplot variables. The package is imported as `iugonet`.

> Some high-resolution cases (induction 20 Hz, ELF 400 Hz, IPRT high-res) use short time ranges to stay light; multi-dimensional model variables (GAIA/Kyushu-GCM 4-D, GPS-TEC 3-D) are listed rather than plotted.

```python
import numpy as np
import iugonet
from pyspedas import tplot, tnames, get_data
print('iugonet', iugonet.__version__)

def show(pattern):
    """List variables matching pattern and tplot the <=2-D ones."""
    vs = tnames(pattern)
    print('created:', vs)
    plottable = []
    for v in vs:
        d = get_data(v)
        if d is None:
            continue
        if np.asarray(d.y).ndim <= 2:
            plottable.append(v)
    if plottable:
        tplot(plottable)
    else:
        print('(multi-dimensional; slice a fixed index to plot)')
```
